```python
"""
    获取股票的基本面数据
    refer to http://tushare.org/fundamental.html
"""
import tushare as ts
import pickle

stock_basics = ts.get_stock_basics()  # 获取股票列表
```

```python
stock_codes = sorted([r[0] for r in stock_basics.iterrows()])
```

```python
def get_stock_data(years, seasons, func):
    data_set = {}
    for y in years:
        for s in seasons:
            year_season_str = '{y}-{s}'.format(y=y,s=s)
            try:
                data_set[year_season_str]=func(y,s)
                print 'get data:', year_season_str
            except:
                print 'fail to get data:', year_season_str
    return data_set
```

```python
fun_map = [
    ('reports', ts.get_report_data),
    ('profits', ts.get_profit_data),
    ('operations', ts.get_operation_data),
    ('growths', ts.get_growth_data), 
    ('debtpayings', ts.get_debtpaying_data), 
    ('cashflows', ts.get_cashflow_data),
]
```

```python
for name, func in fun_map:
    data = get_stock_data([1997+i for i in range(20)], [i+1 for i in range(4)], func)
    with open('tmp_data/stock_{name}_1997-2016.pkl'.format(name=name), 'w') as f:
        pickle.dump(data, f)
```
